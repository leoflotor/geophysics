```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def time(t_min, t_max, dt):
    return np.arange(t_min, t_max + dt, dt)

def wave(frequency, phase, amplitude, time_interval):
    # space = np.arange(0, 2.4, 0.2)
    return amplitude * np.sin(frequency * time_interval + phase)
```

```python
w = np.random.choice(100, 3) / 100
phi = np.random.choice(100, 3) / 100   # initial wave phase
amp = np.random.choice(250, 3) / 100
```

```python
t = time(0, 12 * np.pi, 0.1)
```

```python
wave_data = []
for i in range(0,3):
    wave_data.append(wave(w[i], phi[i], amp[i], t))
```

```python
plt.plot(t, wave_data[0])
plt.plot(t, wave_data[1])
plt.plot(t, wave_data[2]);
```

```python
fig = plt.figure(figsize=(17,11))
ax = fig.gca(projection='3d')
X, Y, Z = wave_data[0], wave_data[1], wave_data[2]

x_plane = [X.min() - 0.5] * len(X)
y_plane = [Y.max() + 0.5] * len(Y)
z_plane = [Z.min() - 0.5] * len(Z)

ax.plot(X, Y, z_plane, label = 'Z projection')
ax.plot(x_plane, Y, Z, label = 'X projection')
ax.plot(X, y_plane, Z, label = 'Y projection')
ax.plot(X, Y, Z, label='Rotation test', marker = 'd', linewidth = 0)

ax.set_xlim(X.min() - 0.5, X.max() + 0.5)
ax.set_ylim(Y.min() - 0.5, Y.max() + 0.5)
ax.set_zlim(Z.min() - 0.5, Z.max() + 0.5)
ax.legend();
```
